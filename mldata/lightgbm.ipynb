```python
import csv
import gc
import json
import pickle
import os

import pandas as pd
import numpy as np
import lightgbm as lgb
```

## Откроем датасет https://www.kaggle.com/datasets/jsphyg/weather-dataset-rattle-package

```python
with open('datasets/weatherAUS.csv') as f:
    reader = csv.reader(f)
    _data = list(reader)
    df = pd.DataFrame(_data[1:], columns=_data[0])
```

```python
df
```

```python
# Берем только числовые колонки
columns = ['RainToday', 'MinTemp', 'MaxTemp', 'Humidity9am', 'Humidity3pm', 'Pressure9am', 'Pressure3pm', 'Temp9am', 'Temp3pm']
# Убираем пустые значения
df = df[df[columns]!='NA'].dropna(subset=columns)
# Делим на X, Y
X, Y = df.loc[:, df.columns!='RainToday'], df.loc[:, df.columns=='RainToday']
df
```

```python
# Приготовим данные, приведем X в float64, а Y в int64
X[columns[1:]] = X[columns[1:]].astype({col: float for col in columns[1:]})
Y = Y.RainToday.map(dict(Yes=1, No=0))
```

```python
gbm = lgb.train(
    {
        'boosting_type': 'gbdt',
        'objective': 'regression',
        'metric': {'l2', 'l1'},
        'num_leaves': 31,
        'learning_rate': 0.05,
        'feature_fraction': 0.9,
        'bagging_fraction': 0.8,
        'bagging_freq': 5,
        'verbose': 0
    },
    lgb.Dataset(X[columns[1:]], Y),
    num_boost_round=20,
    valid_sets=lgb.Dataset(X[columns[1:]], Y),
    callbacks=[lgb.early_stopping(stopping_rounds=5)]
)
```

```python
print(
    gbm.predict(X.loc[:10, columns[1:]], num_iteration=gbm.best_iteration),
    gbm.predict(X.loc[:10, columns[1:]])
)
```

```python
# Сохраняем модель
```

```python
gbm.save_model('models/ligthgbm-binary_cls_model.txt')

with open('models/ligthgbm-binary_cls_model.pckl', 'wb') as f:
    pickle.dump(gbm, f)
```
